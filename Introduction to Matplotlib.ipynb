# Matplotlib example workflow

```python
# 1. Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare Data
x=[1,2,3,4]
y=[11,23,33,44]

# 3. Setup plot
fig,ax=plt.subplots(figsize=(10,10))# parameter is the dimensions of the plot

# 4. Plot Data
ax.plot(x,y)

# 5. Customize Plot
ax.set(title="Simple Plot", xlabel= "x-axis", ylabel= "y-axis")

# 6.Save and show
fig.savefig("sample plot.png")
```

# Multiple plots at the same time(subplots)
## 2 options for subplots

```python
# Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
# Plot to each different axis
ax1.plot(x,x/7);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                     figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randn(1000),np.random.randn(1000));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

#### Which one to use (pyplot vs plt oo method)

* When plotting something quickly its ok to use pyplot
* When plotting something more advanced use plt oo method

```python
over_50=heart_disease[heart_disease['age']>50]
```

```python
over_50
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y='chol',
             c="target",
             ax=ax);
ax.set_xlim([45,100]);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
```

```python
fig,ax=plt.subplots(figsize=(10,6)) # oo method
scatter=ax.scatter(x=over_50["age"],#plot the data
                   y=over_50["chol"],
                   c=over_50["target"])
ax.set(title="Heart disease and Cholesetrol Levels", #customize the plot
       xlabel="Age",
       ylabel="Cholesterol");
#add a legend as the graph is unclear
ax.legend(*scatter.legend_elements(), title= "Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle= '--');
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
#add data to ax0 and ax1
ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"]);

ax1.scatter(x=over_50['age'], y=over_50['thalach'],c=over_50["target"]);


#Customize ax0 and ax1
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

ax1.set(title="Heart Disease and Heart Rate",
        xlabel="Age",
        ylabel="Heart Rate")


#add legends
ax0.legend(*scatter.legend_elements(),title="target")
ax1.legend(*scatter.legend_elements(),title="target")


#add a meanline
ax0.axhline(y=over_50['chol'].mean(),linestyle="--")
ax1.axhline(y=over_50['thalach'].mean(),linestyle="--");

#add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=18, fontweight="bold")
```

```python
fig,ax=plt.subplots(figsize=(10,6)) # oo method
scatter=ax.scatter(x=over_50["age"],#plot the data
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") #change color
ax.set(title="Heart disease and Cholesetrol Levels", #customize the plot
       xlabel="Age",
       ylabel="Cholesterol");
#add a legend as the graph is unclear
ax.legend(*scatter.legend_elements(), title= "Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
#add data to ax0 and ax1
ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"],cmap="summer");

ax1.scatter(x=over_50['age'], y=over_50['thalach'],c=over_50["target"],cmap="winter");


#Customize ax0 and ax1
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

ax1.set(title="Heart Disease and Heart Rate",
        xlabel="Age",
        ylabel="Heart Rate")

#Change x and y axis limits
ax0.set_xlim(50,80)
ax1.set_xlim(50,80)
ax1.set_ylim(80,200)

#add legends
ax0.legend(*scatter.legend_elements(),title="target")
ax1.legend(*scatter.legend_elements(),title="target")


#add a meanline
ax0.axhline(y=over_50['chol'].mean(),linestyle="--",c= "blue")
ax1.axhline(y=over_50['thalach'].mean(),linestyle="--",c="blue");

#add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=18, fontweight="bold")
```
